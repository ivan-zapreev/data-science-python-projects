```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Load the data
file_name = "data" + os.path.sep + "Canada_clean.csv"
df_can = pd.read_csv(file_name)

# Set the Country column to be the index
df_can = df_can.set_index('Country')

# Show the result
df_can.head()
```

```python
# Sort the countries in descending order of the total immigration
df_can.sort_values(by = 'Total', ascending = False, axis = 0, inplace = True)
df_can.head()
```

```python
# Get the top 5 countries data to be plotted
df_top_5 = df_can.iloc[0:5]
display(df_top_5)

# Only take the year colums and transpose the data for plotting
df_top_5 = df_top_5.loc[:,"1980":"2013"].T

# Rename the UK column
df_top_5.rename(columns = {"United Kingdom of Great Britain and Northern Ireland":"United Kingdom"}, inplace = True)

#Show the result
df_top_5.head()
```

```python
# Make the area plot of the data now, do not 
# stack the plots and also set the figure size
df_top_5.plot(kind = "area", 
              stacked = False, 
              alpha=0.25, # 0-1, default value a= 0.5
              figsize = (20,10))

plt.title("The number of immigrants to Canada per a top-5 country")
plt.xlabel("Year")
plt.ylabel("Number of immigrants")
plt.show()
```

```python
# Make the area plot of the data now, do stack 
# the plots and also set the figure size
df_top_5.plot(kind = "area", 
              stacked = True, 
              figsize = (20,10))

plt.title("The number of immigrants to Canada per a top-5 country")
plt.xlabel("Year")
plt.ylabel("Number of immigrants")
plt.show()
```

```python
# Option 2: A preferred option with more flexibility
ax = df_top_5.plot(kind='area', alpha=0.35, figsize=(20, 10))

# Here we operate via the Axes instance
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
# Create a histogram of the immigration numbers in the year 2013 across the countries
df_can['2013'].plot(kind = "hist", color = "red")

plt.title("Distribution of immigrants to Canada in 2013")
plt.xlabel("Immigration numbers")
plt.ylabel("Number of countries")
plt.show()

#The bin edges here are not shown, only the center values
```

```python
# For a better histogram we can compute the bin edges by NumPy library
count, bin_edges = np.histogram(df_can['2013'])

print(type(count))
display(count)
print(type(bin_edges))
display(bin_edges)

# Here we use the computed bin edges for setting the tick marks
df_can['2013'].plot(kind = "hist", xticks = bin_edges,
                    color = "red", figsize=(20, 10))

plt.title("Distribution of immigrants to Canada in 2013")
plt.xlabel("Immigration numbers")
plt.ylabel("Number of countries")
plt.show()
```

```python
# Now let us plot the immigration from Iceland to Canada as a bar chart
years = list(map(str, range(1980, 2014)))

df_iceland = df_can.loc['Iceland', years]

#Make the bar plot
df_iceland.plot(kind = "bar", color = "pink")

plt.title("Immigrants from Iceland to Canada")
plt.xlabel("Year")
plt.ylabel("Number of immigrants")
plt.show()
```

```python
# Let's examine the types of the column labels
all(isinstance(column, str) for column in df_can.columns)

# Set all the columns to strings
df_can.columns = list(map(str, df_can.columns))


display(df_can.columns)
```

```python
# The immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()

# let's get the x-tick values
count, bin_edges = np.histogram(df_t, 15)

# un-stacked histogram
df_t.plot(kind = 'hist', 
          figsize = (10, 6),
          bins = 15,
          alpha = 0.6,
          xticks = bin_edges,
          color = ['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# For a full listing of colors available in Matplotlib, run the following code:
import matplotlib
for name, hex in matplotlib.colors.cnames.items():
    print(name, hex)
```

```python
# If we do no want the plots to overlap each other, we can stack them using the stacked paramemter
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes 
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind = 'hist',
          figsize = (10, 6), 
          bins = 15,
          alpha = 0.6,
          xticks = bin_edges,
          color = ['coral', 'darkslateblue', 'mediumseagreen'],
          stacked = True,
          xlim = (xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

```python
# Display the immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013
# Use an overlapping plot with 15 bins and a transparency value of 0.35.

df_g = df_can.loc[['Greece', 'Albania', 'Bulgaria'], years].transpose()
count, bins = np.histogram(df_g, 15)

ax = df_g.plot(kind = "hist",
         alpha = 0.35,
         xticks = bins,
         bins = 15,
         figsize = (20,10),
         stacked = True)

ax.set_title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
ax.set_ylabel('Number of Years')
ax.set_xlabel('Number of Immigrants') 

plt.show()
```

```python
# Add annotations to a barplot 
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90) # rotate the bars by 90 degrees

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('The fin. crisis',  # s: str. Will leave it blank for no text
             xy = (32, 70),           # place head of the arrow at point (year 2012 , pop 70)
             xytext = (25, 20),       # place base of the arrow at point (year 2008 , pop 20)
             xycoords = 'data',       # will use the coordinate system of the object being annotated 
             arrowprops = dict(
                 arrowstyle = '->', 
                 connectionstyle = 'arc3',
                 color = 'blue', lw = 2)
            )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy = (27, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation = 69.0,                  # based on trial and error to match the arrow
             va = 'bottom',                    # want the text to be vertically 'bottom' aligned
             ha = 'left',                      # want the text to be horizontally 'left' algned.
            )

plt.show()
```

```python
# Using the scripting layter and the df_can dataset, create a horizontal
# bar plot showing the total number of immigrants to Canada from the top
# 15 countries, for the period 1980 - 2013. Label each country with the
# total immigrant count.

df_can.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
df_can.rename(index = {"United Kingdom of Great Britain and Northern Ireland":"United Kingdom"}, inplace = True)
df_top_15 = df_can.head(15)['Total']

display(df_top_15)

ax = df_top_15.plot(kind = "barh",
                   figsize = (20,10))

ax.set_title("Total immigration number for the top 15 countries")
ax.set_xlabel("Country")
ax.set_ylabel("Immigrants")

for idx, value in enumerate(df_top_15):
    # Format using the comma notation
    label = format(int(value), ',')
    # Place on the bar
    plt.annotate(label, xy = (value - 47000, idx - 0.10), color = 'white')
    
plt.show()
```
